```python
import csv

with open('trace_1.DAT') as csvfile:
    data = []
    x_values = []
    y_values = []
    filereader = csv.reader(csvfile, delimiter = ';')
    for line in filereader:
        data.append(line)
    data2 = data[29:]
    #print(data2)
    for lst in data2:
        x_values.append(float(lst[0]))
        y_values.append(float(lst[1]))
    print(x_values)

```

```python
import numpy as np
import matplotlib.pyplot as plt


x_data = np.array(x_values)
y_data = np.array(y_values)
#plt.plot(x_data, y_data)

#print(x_data)
#print(y_data)

plt.scatter(x_data, y_data)
#plt.plot(x_data, y_data)



plt.xlabel("Frequency (Hz)")
plt.ylabel("Power (DB)")
plt.title("Filter Frequency Response")


#plt.grid()

#plt.show()
```

```python
with open('trace_filter.DAT') as csvfile:
    data_filter = []
    x__f_values = []
    y__f_values = []
    filereader = csv.reader(csvfile, delimiter = ';')
    for line in filereader:
        data_filter.append(line)
    data_2 = data_filter[29:]
    #print(data2)
    for lst in data_2:
        x__f_values.append(float(lst[0]))
        y__f_values.append(float(lst[1]))
    print(x__f_values)


```

```python
x__f_data = np.array(x__f_values)
y__f_data = np.array(y__f_values)
#plt.plot(x_data, y_data)

#print(x_data)
#print(y_data)

plt.scatter(x__f_data, y__f_data)
#plt.plot(x_data, y_data)



plt.xlabel("Frequency (Hz)")
plt.ylabel("Power (DB)")
plt.title("Filter Frequency Response")
```
